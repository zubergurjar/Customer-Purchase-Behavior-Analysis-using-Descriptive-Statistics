##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
key_variables = ['Year_Birth', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits',
                  'MntSweetProducts', 'MntWines', 'MntGoldProds']

central_tendency = df[key_variables].agg(['mean', 'median', lambda x: x.mode().iloc[0]], axis=0)
dispersion = df[key_variables].agg(['var', 'std'], axis=0)
```

```python
print("Descriptive Statistics - Central Tendency:")
print(central_tendency)
print("\nDescriptive Statistics - Dispersion:")
print(dispersion)
```

```python
income_outliers = df['Income'][np.abs(df['Income'] - df['Income'].mean()) > 3 * df['Income'].std()]
df['Income'] = np.where(np.abs(df['Income'] - df['Income'].mean()) > 3 * df['Income'].std(),
                        df['Income'].median(), df['Income'])
```

```python
income_outliers
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Income', data=df)
plt.title('Boxplot of Income (Outliers Handled)')
plt.show()
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
total_purchases = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
p_purchase = total_purchases.mean() / total_purchases.max()
expected_purchases = total_purchases.max() * p_purchase
```

```python
from scipy.stats import binom, norm
```

```python
mu_income, sigma_income = df['Income'].mean(), df['Income'].std()
prob_income_gt_70000 = 1 - norm.cdf(70000, mu_income, sigma_income)
expected_income = norm.mean(mu_income, sigma_income)
```

```python
print("Probability Distributions and Calculated Values:")
print(f"Total Purchases - Binomial Distribution:")
print(f"   Probability of a purchase: {p_purchase:.4f}")
print(f"   Expected number of purchases: {expected_purchases:.2f}")

print(f"\nIncome - Normal Distribution:")
print(f"   Probability of income > $70,000: {prob_income_gt_70000:.4f}")
print(f"   Expected income: {expected_income:.2f}")
```

## Task 4: Insights and Customer Segmentation üìà

```python
df['TotalSpend'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
```
